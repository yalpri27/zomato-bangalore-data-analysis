```python
df.rename(columns={'rate (out of 5)': 'ratings'}, inplace=True)
df.rename(columns={'restaurant name': 'restaurant_names'}, inplace=True)
restaurant_names = df['restaurant_names'].value_counts().head(10).index
ratings= df['ratings'].value_counts().head(10).index
```

```python
plt.figure(figsize=(13,6))
plt.bar(restaurant_names, ratings, color='teal')
plt.xlabel("RESTRO NAME")
plt.ylabel("RATINGS OUT OF 5")
plt.title("Top 10 Restaurants (by Rating)")
plt.show()
```
